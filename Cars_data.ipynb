**Obj 1 - **

Find a null value in dataset. If there is any null value in any column, then process it.

```python
cars.shape
```

```python
cars.describe()
```

```python
cars.isnull().sum()
```

```python
cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace=True)
```

```python
cars.dtypes
```

```python
cars['EngineSize'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['Horsepower'].fillna(cars['Cylinders'].mean(), inplace=True)
```

```python
cars['MPG_City'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['MPG_Highway'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['Weight'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['Wheelbase'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['Length'].fillna(cars['Cylinders'].mean(), inplace=True)
```

```python
cars.isnull().sum()
```

```python
cars.dropna(inplace=True)
```
